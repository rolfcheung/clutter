```python
%matplotlib inline
from IPython.display import set_matplotlib_formats
set_matplotlib_formats('retina')

# import packages
import pandas as pd
import matplotlib.pyplot as plt
import os
import numpy as np
```

```python
# read data
data = pd.read_csv('../data/carbon_intensity_per_captia.csv', index_col=0)

data
```

```python
city_name = list(data.head(0))
```

```python
fig = plt.figure(figsize=(40,60))

x = 0.05
y = 0.9
fontsize = 40
fontweight = 'heavy'
width = 0.83
yticks_fontsize = 30

for i in range(0, 7):
    subset = data.iloc[:,i]
    ax = fig.add_subplot(8, 4, i+1)
    subset.plot.bar(ax = ax, 
                    legend=False,
                    color ='#ca0020',
                    width = 0.83)
    plt.text(x, y,city_name[i], 
         {'color': '#ca0020', 'fontsize': fontsize, 'name': 'Times New Roman'},
         transform=ax.transAxes)
    plt.yticks(fontsize=yticks_fontsize)
    plt.xticks(())
    
    
for i in range(7, 16):
    subset = data.iloc[:,i]
    ax = fig.add_subplot(8, 4, i+1)
    subset.plot.bar(ax = ax, 
                    legend=False,
                    color ='#f4a582',
                    width = 0.83)
    plt.text(x, y,city_name[i], 
         {'color': '#f4a582', 'fontsize': fontsize, 'name': 'Times New Roman'},
         transform=ax.transAxes)
    plt.yticks(fontsize=yticks_fontsize)
    plt.xticks(())
    

for i in range(16, 26):
    subset = data.iloc[:,i]
    ax = fig.add_subplot(8, 4, i+1)
    subset.plot.bar(ax = ax, 
                    legend=False,
                    color ='#35978f',
                    width = 0.83)
    plt.text(x, y,city_name[i], 
         {'color': '#35978f', 'fontsize': fontsize, 'name': 'Times New Roman'},
         transform=ax.transAxes)
    plt.yticks(fontsize=yticks_fontsize)
    plt.xticks(())
    
    
for i in range(26, 29):
    subset = data.iloc[:,i]
    ax = fig.add_subplot(8, 4, i+1)
    subset.plot.bar(ax = ax, 
                    legend=False,
                    color ='#0571b0',
                    width = 0.83)
    plt.text(x, y,city_name[i], 
         {'color': '#0571b0', 'fontsize': fontsize, 'name': 'Times New Roman'},
         transform=ax.transAxes)
    plt.yticks(fontsize=yticks_fontsize)
    plt.xticks(())
    
    
    
fig.text(0.325, 0.06,'Anyang: 7 cities with a trend of abrupt fluctuation', 
         {'color': '#ca0020', 'fontsize': fontsize, 'name': 'Times New Roman'})

fig.text(0.325, 0.05,'Jiyuan: 9 cities with a trend of steady upward', 
         {'color': '#f4a582', 'fontsize': fontsize, 'name': 'Times New Roman'})

fig.text(0.325, 0.04,'Jiaozuo: 10 cities with a trend of rising at first and decreasing then', 
         {'color': '#92c5de', 'fontsize': fontsize, 'name': 'Times New Roman'})

fig.text(0.325, 0.03,'Puyang: 3 cities with a trend of relative stability', 
         {'color': '#0571b0', 'fontsize': fontsize, 'name': 'Times New Roman'})

fig.text(0.325, 0.02,'Time span: 2000--2015', 
         {'fontsize': fontsize, 'name': 'Times New Roman'})



fig.tight_layout()



fig.tight_layout()

plt.savefig('../result/carbon_intensity_per_capita.svg',dpi=400, alpha=True)
plt.show()
```
